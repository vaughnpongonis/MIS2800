## Create Table

### Subtask:
Create a new SQL table named 'products' with three columns: 'product_id' (INTEGER PRIMARY KEY), 'name' (TEXT), and 'price' (REAL), using STRICT typing.


```python
conn = sqlite3.connect('products.db')
print("products.db database created")
%sql sqlite:///products.db
```

**Reasoning**:
Now that the database `products.db` is created and connected, the next step is to create the `products` table within this database, as specified in the subtask. This involves defining the columns `product_id` (INTEGER PRIMARY KEY), `name` (TEXT), and `price` (REAL), and ensuring `STRICT` typing.



```python
%%sql
DROP TABLE IF EXISTS products;
CREATE TABLE products (
    product_id INTEGER PRIMARY KEY,
    name TEXT,
    price REAL
) STRICT;
```

## Insert Data

### Subtask:
Insert at least three items into the 'products' table, including 'Superbowl Jersey' as an example and two additional products.


**Reasoning**:
Now that the 'products' table has been created, I need to insert the specified data into it. This will involve three `INSERT INTO` statements within a single `%%sql` block, ensuring 'Superbowl Jersey' is one of the entries.



```python
%%sql
INSERT INTO products (product_id, name, price) VALUES (1, 'Superbowl Jersey', 75.00);
INSERT INTO products (product_id, name, price) VALUES (2, 'Official NFL Football', 49.99);
INSERT INTO products (product_id, name, price) VALUES (3, 'Team Hat', 25.50);
```

```python
%%sql
SELECT * FROM products WHERE price > 50;
```

```python
%%sql
UPDATE products
SET price = 99.99
WHERE name = 'Superbowl Jersey';
```
